```python
keys_col = {}

for col in df.columns:
    keys_col[col] = df[col].dtype.name
    
keys_col
```

```python
for col in keys_col:
    if keys_col[col] == 'float64':
        plt.figure(figsize=(15, 2))
        sns.boxplot(df[col])
        plt.title(f'Boxplot for {col}')
        plt.show()
```

```python
f, (ax_box, ax_hist) = plt.subplots(2, 
                                    sharex=True, 
                                    gridspec_kw={"height_ratios": (.15, .85)}, 
                                    figsize=(15,5))

sns.distplot(df['relative_velocity'], ax=ax_hist)
sns.boxplot(df['relative_velocity'], ax=ax_box)

plt.show()
```

```python
g = sns.JointGrid(height=8)

sns.distplot(df['relative_velocity'], ax=g.ax_joint, bins=25)
sns.boxplot(df['relative_velocity'], ax=g.ax_marg_x)
plt.show()
```
